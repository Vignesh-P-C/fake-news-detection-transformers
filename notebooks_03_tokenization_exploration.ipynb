```python
fake_df = pd.read_csv(
    "Fake.csv",
    engine="python",
    sep=",",
    quotechar='"',
    escapechar="\\",
    on_bad_lines="skip"
)

texts = fake_df["text"].sample(5, random_state=42).tolist()
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
example = texts[0]
tokens = tokenizer(example)

len(tokens["input_ids"])
```
